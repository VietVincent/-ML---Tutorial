```
#@One hidden layer
class Net1(nn.Module):
  def __init__(self, in_size, out_size):
    super(Net1, self).__init__()
    self.linear = nn.Linear(in_size, out_size, bias=True)
  def forward(self, x):
    out = self.linear(x)
    return out
```

```
#@One hidden layer + Sigmoid
class Net2(nn.Module):
  def __init__(self, in_size, hidden_size, out_size):
    super(Net2, self).__init__()
    self.linear1 = nn.Linear(in_size, hidden_size, bias=True)
    self.act1 = nn.Sigmoid()
    self.linear2 = nn.Linear(hidden_size, out_size,bias=True)
  def forward(self, x):
    out = self.linear1(x)
    out = self.act1(out)
    out = self.linear2(out)
    return out
```

```
#@One hidden layer + ReLU
class Net3(nn.Module):
  def __init__(self, in_size, hidden_size, out_size):
    super(Net3, self).__init__()
    self.linear1 = nn.Linear(in_size, hidden_size, bias=True)
    self.act1 = nn.ReLU()
    self.linear2 = nn.Linear(hidden_size,out_size,bias=True)
  def forward(self, x):
    out = self.linear1(x)
    out = self.act1(out)
    out = self.linear2(out)
    return out
```

```
#@Two hidden layer + ReLU
class Net4(nn.Module):
  def __init__(self, in_size, hidden_size1, hidden_size2, out_size):
    super(Net4, self).__init__()
    self.linear1 = nn.Linear(in_size, hidden_size1, bias=True)
    self.act1 = nn.ReLU()
    self.linear2 = nn.Linear(hidden_size,out_size,bias=True)
  def forward(self, x):
    out = self.linear1(x)
    out = self.act1(out)
    out = self.linear2(out)
    return out
```
